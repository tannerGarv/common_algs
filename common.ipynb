```python
import math
'''
   Sphere volume formula: 4/3 * pi * radius ^ 3
   
   The above is the formula for the volume of a sphere.
   Write a functin that returns the volume of a sphere
   given a radius 'r' as input.
'''
def sphere_Volume(radius):
   return (4/3) * math.pi * (radius**3) # (radius**3) same as math.pow(radius, 3)

print(sphere_Volume(3))
```

```python
'''
   Write a function that plays the FizzBuzz game with the numbers 1-100, inclusive.
   This is a well-known intro task to get you comfortable with basic control flow logic.
   
   Print the numbers 1-100, replacing all numbers divisible by 3 with Fizz.
'''
import math
def fizz_buzz_game():
   for i in range(1, 101):
      if i % 3 == 0 and i % 5 == 0:
         print('FizzBuzz')
      elif i % 5 == 0: 
         print('Buzz')
      elif i % 3 == 0:
         print('Fizz')
      else:
         print(i)
      
fizz_buzz_game()
```

```python
'''
   Palindrome: A phrase that coverts all uppercase letters into lowercase letters and removing all non-alphabetic characters,
   it reads the same backwards and forwards.
'''
import math
def is_palindrome(s):
   sanitized_s = "".join(char for char in s if char.isalpha()).lower()  # deletes spaces, removes numbers
   left_ptr, right_ptr = 0, len(sanitized_s)-1
   while left_ptr < right_ptr:
      if sanitized_s[left_ptr] != sanitized_s[right_ptr]:
         return False
      left_ptr += 1
      right_ptr -= 1
   return True

print(is_palindrome("A ma524n, a5 plan, a can532al, Panama!")) # true
print(is_palindrome("Was it a car or a cat I saw?")) # true
print(is_palindrome("Palindrome")) # false
```

```python
def gcd(a,b):
   
   if a == 0:
      q = b
      return print(f'GCD is _{b}_ : A is at zero')
   
   if b == 0:
      q = a
      return print(f'GCD is _{a}_ : B is at zero')
   
   if a > b :
      q = a / b      # get the quotient
      r = a % b      # get the remainder
      if r == 0:
         print(f'GCD is _{q}_ : remainder is zero')
      else:
         gcd(b,r) # recursive call if no GCD
         
print(gcd(96,5))
      
```
